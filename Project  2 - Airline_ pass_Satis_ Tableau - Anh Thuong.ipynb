# 5.  Từ cột "Total delay" tạo bin = 30 phút, tính tỉ lệ số lượng khách hàng hài lòng/không hài lòng (cột "satisfaction") cho 5 bin đầu tiên? Vẽ biểu đồ minh họa và nhận xét kết quả.


```python
# nếu bỏ hết ouliers của Total delay thì max = 60 -> ko thể chia nhóm từ 60 trở lên -> giữ lại outlier
# Giữ nguyên outlier

# chia nhóm
bins = [0, 30, 60, 90, 120, 150]
labels = ['[0,30]', '[30,60]', '[60,90]', '[90,120]', '[120,150]']

airline['Delay_group'] = pd.cut(airline['Total delay'], bins=bins, labels=labels, include_lowest=True)

#  Tính số lượng khách hàng theo mức độ hài lòng
delay_satis = airline.groupby(['Delay_group', 'satisfaction']).size().unstack(fill_value=0)

# set thứ tự cột
delay_satis = delay_satis.reindex(columns=['satisfied', 'neutral or dissatisfied'])

# Tính phần trăm từng nhóm
delay_satis_pct = delay_satis.div(delay_satis.sum(axis=1), axis=0) * 100

# Vẽ biểu đồ stacked bar
result = delay_satis_pct.plot(
    kind='bar',
    stacked=True,
    color=['blue', 'orange'],
    figsize=(8,4)
)

plt.title('Tỷ lệ % khách hàng hài lòng theo từng mức độ trễ')
plt.xlabel('Khoảng thời gian trễ (phút)')
plt.ylabel('Tỷ lệ %')
plt.legend(title='', loc='upper right')
plt.xticks(rotation= 0)
plt.tight_layout()
plt.show()
```

**Nhận xét:**
- Khoảng thời gian thấp (0,30) thì mức độ hài lòng cao nhất (đến 45%), khoảng thời gian trễ càng tăng lên thì mức độ hài lòng giảm dần
- Đối với khoảng thời gian trễ từ (120,150) thì độ hài lòng giảm còn 35-38%.
- Mối quan hệ thời gian trễ càng lâu → mức độ hài lòng càng giảm.
=> Cần ưu tiên giảm thiểu thời gian trễ chuyến, đặc biệt là các chuyến bay có khả năng trễ >60 phút.
